```python
import pandas as pd
from pandas.core.frame import DataFrame
import aiohttp
import asyncio
import json
from tqdm import tqdm
from typing import Dict
import os
import aiohttp
import matplotlib.pyplot as plt

TIMEOUT = aiohttp.ClientTimeout(
    total=None
)

def save_results(results, filename):
    with open(filename, "w") as f:
        json.dump(results, f, indent=4)

async def run_inference(model_name, prompt, temperature=0.0, max_tokens=16):
    """
    curl http://localhost:11434/api/generate -d '{
		"model": "llama3.2",
		"prompt":"Why is the sky blue?",
		"stream": false,
	}'
    """
    async with aiohttp.ClientSession(timeout = TIMEOUT) as session:
        async with session.post(
            "http://localhost:11434/api/generate",
            json={
                "model": model_name,
                "prompt": prompt,
                "stream": False,
                "options": {
                    "temperature": temperature,
                    "num_predict": max_tokens,
                },
            }
        ) as response:
            result = await response.json()
            return result.get("response", "")

async def run_batch_inference(model_name, prompts, temperature=0.0, max_tokens=16):
    progress = tqdm(total=len(prompts))
    async def _run_inference(prompt):
        result = await run_inference(model_name, prompt, temperature, max_tokens)
        progress.update(1)
        return result
    tasks = [_run_inference(prompt) for prompt in prompts]
    return await asyncio.gather(*tasks)

def cache_inferences(model_name, exp_name, prompts, inferences):
    if os.path.exists(f"inference_cache/{model_name}_{exp_name}.json"):
        with open(f"inference_cache/{model_name}_{exp_name}.json", "r") as f:
            cache = json.load(f)
            cache.update(dict(zip(prompts, inferences)))
    else:
        cache = dict(zip(prompts, inferences))
    with open(f"inference_cache/{model_name}_{exp_name}.json", "w") as f:
        json.dump(cache, f)

def load_inferences(model_name, exp_name, prompts):
    if os.path.exists(f"inference_cache/{model_name}_{exp_name}.json"):
        with open(f"inference_cache/{model_name}_{exp_name}.json", "r") as f:
            cache = json.load(f)
            return [cache[prompt] if prompt in cache else None for prompt in prompts]
    return [None for _ in prompts]

def get_remaining_prompts(model_name, exp_name, prompts):
    if os.path.exists(f"inference_cache/{model_name}_{exp_name}.json"):
        with open(f"inference_cache/{model_name}_{exp_name}.json", "r") as f:
            cache = json.load(f)
            return [prompt for prompt in prompts if prompt not in cache]
    return prompts

models = [
    "llama3.2:1b" ,
    "llama3.2",
    "gemma3:1b",
    "gemma3:4b",
]
```

### Part 1.1: Confusion Matrix (5 points)
Complete the following functions. The first one should computes a confusion matrix for a set of predictions, and the second one should use your favorite tool (e.g. Matplotlib) to visualize it as a heatmap. 

We're implementing the confusion matrix first because you might find it useful when completing the rest of Part 1.

If the model messes up the output format (e.g. outputs something that isn't one of the labels), the prediction should not be included in the matrix counts.

```python
TRUE_COL = "newsgroup"
import seaborn as sns
def part1_1_cm(x: DataFrame, y: str) -> Dict[str, Dict[str, int]]:
    col = y if y in x.columns else f"predictions_{y}"
    
    gt = x[TRUE_COL].astype(str).str.strip()
    pred = x[col].astype(str).str.strip()
    valid = pred.isin(set(gt.unique()))  # only include valid predictions
    
    matrix: Dict[str, Dict[str, int]] = {}
    for t, p in zip(gt[valid], pred[valid]):
        if t not in matrix:
            matrix[t] = {}
        if p not in matrix[t]:
            matrix[t][p] = 0
        matrix[t][p] += 1
    
    return matrix

def part_1_1_vis(x: Dict[str, Dict[str, int]]):
    df = pd.DataFrame.from_dict(x).fillna(0).astype(int)
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(df, annot=False, fmt="d", cmap="Blues", cbar=True)
    plt.xlabel("True Label")
    plt.ylabel("Predicted Label")
    plt.title("Confusion Matrix")
    plt.show()
```

Now we'll use these functions to compute and visualize confusion matrices for each classifier.

```python
confusion_matrices = dict()

for model in models:
    confusion_matrices[model] = part1_1_cm(classification_data, model)
    part_1_1_vis(confusion_matrices[model])
```

```python
# TODO: Use this scratch space to find documents 
# that are examples of common misclassfications
from typing import List, Tuple, Dict
def top_confusions(cm: Dict[str, Dict[str, int]], k=5) -> List[Tuple[str, str, int]]:
    pairs = []
    for t, row in cm.items():
        for p, c in row.items():
            if t != p and c > 0:
                pairs.append((t, p, c))
    return sorted(pairs, key=lambda x: x[2], reverse=True)[:k]

def example_misclassifications(df: pd.DataFrame, model_col: str, true_label: str, pred_label: str, n=3) -> pd.DataFrame:
    col = model_col if model_col in df.columns else f"predictions_{model_col}"
    gt = df[TRUE_COL].astype(str).str.strip()
    pred = df[col].astype(str).str.strip()
    mask = (gt == true_label) & (pred == pred_label)
    out = df.loc[mask, ["id", TRUE_COL, col, "text"]].head(n).copy()
    out["snippet"] = out["text"].astype(str).str.replace(r"\s+", " ", regex=True).str[:200] + "..."
    return out


for m in models:
    cm = confusion_matrices[m]
    print(f"\nTop confusions for {m}:")
    for t, p, c in top_confusions(cm, k=3):
        print(f"  {t} → {p} : {c} docs")
        ex = example_misclassifications(classification_data, m, t, p, n=2)
        for _, row in ex.iterrows():
            print(f"   [id={row['id']}] {row['snippet']}")
```

```python
TRUE_COL = "newsgroup"

def part1_2_acc(x: pd.DataFrame, y: str) -> float:
    """
    Accuracy = (# correct predictions) / (total predictions).
    """
    col = y if y in x.columns else f"predictions_{y}"
    gt   = x[TRUE_COL].astype(str).str.strip()
    pred = x[col].astype(str).str.strip()
    
    correct = (pred == gt).sum()
    total   = len(pred)
    
    return correct / total if total > 0 else 0.0

accuracies: dict[str, float] = dict()

for model in models:
    accuracies[model] = part1_2_acc(classification_data, model)

save_results(accuracies, "results/accuracies.json")

best_model = "gemma3:4b" # TODO: Which model performed the best?
```

### Part 1.3: Precision (5 points)
Complete the following function that returns, for each class, the precision of the classifier's labels given ground-truth labels. (note: if there are no predictions for a given class, then precision should be 0)

★ Generate a table showing, for the best-performing classifier, the precision of its labels across each class.

Table for the best performing model is displayed below.

★ For which class are its labels the most accurate? The least?

The class for which the labels are most accurate are the rec.sports.hockey column and also the comp.windows.x and ones that are least accurate include the political categories.

```python
def part1_3_prec(x: DataFrame, y: str) -> dict[str, float]:
    """
    Compute per-class precision for classifier predictions vs ground truth.
    Precision(c) = TP(c) / (TP(c) + FP(c)), 0 if denominator=0.
    """
    col = y if y in x.columns else f"predictions_{y}"
    gt   = x[TRUE_COL].astype(str).str.strip()
    pred = x[col].astype(str).str.strip()
    
    labels = sorted(gt.unique())
    precision: dict[str, float] = {}
    
    for c in labels:
        tp = ((pred == c) & (gt == c)).sum()
        fp = ((pred == c) & (gt != c)).sum()
        denom = tp + fp
        precision[c] = tp / denom if denom > 0 else 0.0
    
    return precision
```

### Part 1.4: Recall (5 points)
Complete the following function that returns, for each target class, the recall of the classifier's labels. (note: if there are no ground truth labels for a given class, the recall should be 0)

```python
def part1_4_rec(x: DataFrame, y: str) -> dict[str, float]:
    """
    Compute per-class recall for classifier predictions vs ground truth.
    Recall(c) = TP(c) / (TP(c) + FN(c)), 0 if no ground-truth instances for c.
    """
    col  = y if y in x.columns else f"predictions_{y}"
    gt   = x[TRUE_COL].astype(str).str.strip()
    pred = x[col].astype(str).str.strip()

    labels = sorted(gt.unique())
    recall: dict[str, float] = {}

    for c in labels:
        tp = ((pred == c) & (gt == c)).sum()
        fn = ((pred != c) & (gt == c)).sum()
        denom = tp + fn
        recall[c] = tp / denom if denom > 0 else 0.0

    return recall

best_model_rec = part1_4_rec(classification_data, best_model)

save_results(best_model_rec, "results/best_model_rec.json")

# TODO: Scratch space for analyzing the recall.
```

### Part 1.5: F1 (5 points)

A weighted average F1 score computes an average F1 score over all target classes, weighted by the true number of instances of that class:

![test](A1_Data/f1.png)

Here, *f* is a classification function, *C* is the set of classes, *D* is the set of evaluation data with ground-truth labels, and *D_c* is the set of data whose ground-truth label is *c*.

Implement two functions below:
* A function that returns, for a target class, the F1 of a classifier's labels given ground truth labels
* A function that uses the previous function to compute a weighted average F1 score

```python
import pandas as pd

TRUE_COL = "newsgroup"

def part1_5_class_f1(x: pd.DataFrame, y: str, z: str) -> float:
    """
    Per-class F1 for class z.
    F1(z) = 2*TP / (2*TP + FP + FN); returns 0 if denominator = 0.
    """
    col  = y if y in x.columns else f"predictions_{y}"
    gt   = x[TRUE_COL].astype(str).str.strip()
    pred = x[col].astype(str).str.strip()

    tp = ((pred == z) & (gt == z)).sum()
    fp = ((pred == z) & (gt != z)).sum()
    fn = ((pred != z) & (gt == z)).sum()

    denom = 2*tp + fp + fn
    return (2*tp / denom) if denom > 0 else 0.0

def part1_5_weighted_f1(x: pd.DataFrame, y: str) -> float:
    """
    Weighted F1 across classes, weighting by true class frequency.
    \bar{F1} = sum_c (|D_c|/|D|) * F1(c)
    """
    col  = _resolve_pred_col(x, y)
    gt   = x[TRUE_COL].astype(str).str.strip()
    pred = x[col].astype(str).str.strip()

    labels = sorted(gt.unique())
    N = len(gt)
    if N == 0:
        return 0.0

    wsum = 0.0
    for c in labels:
        w = (gt == c).sum() / N          # |D_c| / |D|
        # compute F1 for c directly (avoids re-scanning all rows twice)
        tp = ((pred == c) & (gt == c)).sum()
        fp = ((pred == c) & (gt != c)).sum()
        fn = ((pred != c) & (gt == c)).sum()
        denom = 2*tp + fp + fn
        f1_c = (2*tp / denom) if denom > 0 else 0.0
        wsum += w * f1_c
    return wsum
```

### Part 2.1: Word error rate (20 points)
Implement the following function to compute the word error rate for a candidate transcription given a reference. Each of the recordings we transcribed are readings of the same English text: *"Please call Stella.  Ask her to bring these things with her from the store:  Six spoons of fresh snow peas, five thick slabs of blue cheese, and maybe a snack for her brother Bob.  We also need a small plastic snake and a big toy frog for the kids.  She can scoop these things into three red bags, and we will go meet her Wednesday at the train station."*

```python
def part2_1(x: list[str], y: list[str]) -> float:
    """
    Compute Word Error Rate (WER) between reference (x) and candidate (y).
    
    Args:
        x: list of tokens (reference transcription)
        y: list of tokens (candidate transcription)
    
    Returns:
        WER as float (0.0–inf, can exceed 1.0 if many insertions).
    """
    n = len(x)
    
    # Create edit distance matrix
    dp = np.zeros((len(x)+1, len(y)+1), dtype=int)
    for i in range(len(x)+1):
        dp[i][0] = i
    for j in range(len(y)+1):
        dp[0][j] = j
    
    # Fill matrix
    for i in range(1, len(x)+1):
        for j in range(1, len(y)+1):
            if x[i-1] == y[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                substitution = dp[i-1][j-1] + 1
                insertion    = dp[i][j-1] + 1
                deletion     = dp[i-1][j] + 1
                dp[i][j] = min(substitution, insertion, deletion)
    
    # Edit distance = bottom-right cell
    edit_distance = dp[len(x)][len(y)]
    
    return edit_distance / n if n > 0 else 0.0
```

### Part 3.1 n-gram precision (5 points)

complete the following function to calculate the n-gram prediction of a predicted translation given the reference.

```python

# def ng_prec(pred, ref, n):
#     # TODO: Implement this function.
#     pass

# from collections import Counter

# def ngrams(tokens, n):
#     """Return list of n-grams from a list of tokens."""
#     return [tuple(tokens[i:i+n]) for i in range(len(tokens)-n+1)]

# def ng_prec(pred, ref, n):
#     """
#     Compute n-gram precision of a predicted translation against reference.

#     Args:
#         pred (str): predicted translation
#         ref (str): reference translation
#         n (int): n-gram size
    
#     Returns:
#         float: n-gram precision
#     """
#     pred_tokens = pred.strip().split()
#     ref_tokens = ref.strip().split()
    
#     pred_ngrams = Counter(ngrams(pred_tokens, n))
#     ref_ngrams  = Counter(ngrams(ref_tokens, n))
    
#     # Clipped count
#     overlap = {ng: min(count, ref_ngrams[ng]) 
#                for ng, count in pred_ngrams.items()}
    
#     numerator = sum(overlap.values())
#     denominator = sum(pred_ngrams.values())
    
#     if denominator == 0:
#         return 0.0
    
#     return numerator / denominator

def _ngrams(tokens, n):
    return [tuple(tokens[i:i+n]) for i in range(len(tokens)-n+1)]

def _tok_list(x):
    # Return a list of token lists
    if isinstance(x, str):
        return [x.strip().split()]
    if isinstance(x, (list, tuple)):
        if not x:
            return []
        # list of tokens (single example)
        if all(isinstance(t, str) and (' ' not in t) for t in x):
            return [list(x)]
        # list of sentences (strings)
        return [s.strip().split() for s in x]
    raise TypeError("pred/ref must be str or list/tuple")

def _corpus_ng_prec(preds, refs, n):
    """Clipped n-gram precision over a whole corpus."""
    pred_counts = Counter()
    ref_counts  = Counter()
    for p_toks, r_toks in zip(_tok_list(preds), _tok_list(refs)):
        pred_counts.update(_ngrams(p_toks, n))
        ref_counts.update(_ngrams(r_toks, n))
    if not pred_counts:
        return 0.0
    overlap = sum(min(cnt, ref_counts[ng]) for ng, cnt in pred_counts.items())
    return overlap / sum(pred_counts.values())

def ng_prec(pred, ref, n):
    """Sentence-level clipped precision (kept for Part 3.1)."""
    p = _tok_list(pred)[0]
    r = _tok_list(ref)[0]
    if len(p) < n:
        return 0.0
    from collections import Counter
    p_ngrams = Counter(_ngrams(p, n))
    r_ngrams = Counter(_ngrams(r, n))
    overlap = sum(min(cnt, r_ngrams[ng]) for ng, cnt in p_ngrams.items())
    return overlap / max(1, sum(p_ngrams.values()))
```

```python
# def brev_pen(pred, ref):
#     # TODO: Implement this function.
#     pass

from math import exp

# def brev_pen(pred, ref):
#     """
#     BLEU brevity penalty.
#     BP = 1                      if c > r
#          exp(1 - r / c)        otherwise
#     where c is candidate length, r is reference length (in tokens).

#     Accepts:
#       - pred/ref as strings (single sentence), or
#       - lists of strings (corpus); sums token counts.
#       - If given lists of tokens, treats length directly.
#     """
#     def tok_len(x):
#         if isinstance(x, str):
#             return len(x.strip().split())
#         if isinstance(x, (list, tuple)):
#             if not x:
#                 return 0
#             # list of tokens
#             if all(isinstance(t, str) and (' ' not in t) for t in x):
#                 return len(x)
#             # list of sentences
#             return sum(len(s.strip().split()) for s in x)
#         raise TypeError("pred/ref must be str or list/tuple")

#     c = tok_len(pred)
#     r = tok_len(ref)

#     if c == 0:
#         return 0.0
#     if c > r:
#         return 1.0
#     return exp(1 - r / c)


def brev_pen(pred, ref):
    """BP computed on total corpus lengths if lists are given."""
    def len_tokens(x):
        if isinstance(x, str):
            return len(x.strip().split())
        if isinstance(x, (list, tuple)):
            if not x:
                return 0
            # list of tokens
            if all(isinstance(t, str) and (' ' not in t) for t in x):
                return len(x)
            # list of sentences
            return sum(len(s.strip().split()) for s in x)
        raise TypeError("pred/ref must be str or list/tuple")
    c = len_tokens(pred)
    r = len_tokens(ref)
    if c == 0:
        return 0.0
    if c > r:
        return 1.0
    return math.exp(1 - r / c)
```

### Part 3.3 putting it together (10 points)

complete the following function to calculate the full bleu score, using the two functions from the previous parts. Represent the bleu score as discussed in class, using the 1, 2, 3, and 4 gram precisions.

```python
# def bleu(pred, ref):
#     # TODO: Implement this function.
#     pass

import math

# def bleu(pred, ref):
#     """
#     Compute BLEU-4 score between predicted and reference translations.
#     Uses n-gram precisions for n=1..4 and applies brevity penalty.
#     """
#     precisions = []
#     for n in range(1, 5):
#         p_n = ng_prec(pred, ref, n)
#         if p_n == 0:
#             return 0.0  # if any precision is zero, BLEU=0
#         precisions.append(p_n)

#     # geometric mean of log precisions
#     log_prec = sum(math.log(p) for p in precisions) / 4.0
#     geo_mean = math.exp(log_prec)

#     # brevity penalty
#     bp = brev_pen(pred, ref)

#     return bp * geo_mean

def bleu(pred, ref):
    """
    BLEU-4.
    - If `pred`/`ref` are strings: sentence BLEU.
    - If lists of strings: **corpus BLEU** (what the tests expect).
    No smoothing; equal weights for n=1..4.
    """
    is_corpus = isinstance(pred, (list, tuple)) or isinstance(ref, (list, tuple))

    if is_corpus:
        precisions = []
        for n in range(1, 5):
            p_n = _corpus_ng_prec(pred, ref, n)
            if p_n == 0:
                return 0.0
            precisions.append(p_n)
        geo_mean = math.exp(sum(math.log(p) for p in precisions) / 4.0)
        bp = brev_pen(pred, ref)
        return bp * geo_mean

    # sentence-level fallback
    precisions = []
    for n in range(1, 5):
        p_n = ng_prec(pred, ref, n)
        if p_n == 0:
            return 0.0
        precisions.append(p_n)
    geo_mean = math.exp(sum(math.log(p) for p in precisions) / 4.0)
    bp = brev_pen(pred, ref)
    return bp * geo_mean

```

```python
judge_model = "llama3.2:1b"

def get_pairwise_judgement_prompt(output1, output2):
	prompt = f"Which of the following two outputs is better? Output 1: {output1}. Output 2: {output2}. Output only the number 1 or 2, no other text."
	return prompt

head_to_head_judgements = [{str((model_a, model_b)): None for model_a in models for model_b in models if model_a != model_b} for _ in range(len(prompts))]

import itertools, re

# --- parser: accept "2", "Output 2", "I choose 1", etc. ---
_digit = re.compile(r"\b([12])\b")
def _parse_vote(s: str) -> int:
    if isinstance(s, str):
        m = _digit.search(s.strip())
        if m:
            return int(m.group(1))
    return 0  # 0 = invalid/unclear

async def run_head_to_head():
    P = len(prompts)
    total, nonempty = 0, 0

    # ordered pairs because your keys are ordered (a,b)
    for a, b in itertools.permutations(models, 2):
        for i in range(P):
            prompt = get_pairwise_judgement_prompt(
                outputs_per_model[a][i],
                outputs_per_model[b][i],
            )
            txt = await run_inference(
                judge_model, prompt,
                temperature=0.0,
                max_tokens=1024*8   # important for gpt-oss / long-think models
            )

            vote = _parse_vote(txt)
            head_to_head_judgements[i][str((a, b))] = vote

            total += 1
            if isinstance(txt, str) and txt.strip():
                nonempty += 1

    print(f"[judge] ran {total} evaluations; non-empty responses: {nonempty}/{total}")
    return head_to_head_judgements

# ---- run and save ----
head_to_head_judgements = await run_head_to_head()
save_results(head_to_head_judgements, "results/head_to_head_judgements.json")

```
